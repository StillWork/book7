## 키-몸무게-운동선수

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

n_samples = 200
x1 = 5*np.random.randn(n_samples) + 190 #  농구선수
x2 = 4*np.random.randn(n_samples) + 180 #  축구선수

z1 = 4*np.random.randn(n_samples) + 80 # 몸무게
z2 = 4*np.random.randn(n_samples) + 70
plt.hist(x1, bins=30)
plt.hist(x2, bins=30)

plt.hist(z1, bins=30)
plt.hist(z2, bins=30)
plt.show()
```

```python
n_samples*[0]
```

```python
x = np.concatenate((x1, x2)) # 키
z = np.concatenate((z1, z2))  # 몸무게
y = n_samples*[1] + n_samples*[0] # 0: 농구, 1:축구
```

```python
# ctrl-m-b
z
```

```python
import pandas as pd
df = pd.DataFrame([x,z,y]).T
df.columns = ['키','몸무게','종목']
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,z,s=15,c=y) # y는 레이블, 실제값, ground truth
```

```python
from sklearn.cluster import KMeans
X = df[['키','몸무게']]
model = KMeans(n_clusters=2).fit(X)
# model.labels_
```

```python
model.labels_
```

```python
model.predict([[190,79]])
```

```python
model.fit_predict(X)
```

```python
plt.scatter(x,z,c=model.fit_transform(X))
```

```python
(model.predict(X)==y).mean()
```
